```python
import pandas as pd
import cutlet
katsu = cutlet.Cutlet()
```

```python
import os
dir_string = r"C:\Users\arneg\Documents\arbeit\GitHub\BerlinLab\automatic_transcription\jap_data"
for subdir, dirs, files in os.walk(dir_string):
            for file in files:
                if file.endswith('.xlsx'):
                    
                    file_path = os.path.join(dir_string,subdir,file)
                    df = pd.read_excel(file_path)
                    for i in range(len(df)):
                        if pd.isna(df.loc[i,'automatic_transcription']):
                            continue
                        else:
                            rom = katsu.romaji(df.loc[i,'automatic_transcription'])
                            df.loc[i,'automatic_transcription'] = df.loc[i,'automatic_transcription'] + " - " + rom 
                    df.to_excel(file_path)
```
